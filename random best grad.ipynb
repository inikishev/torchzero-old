# SGD + nesterov

```python
import torch
from torchzero.optim import RandomGrad, RandomBestGrad
from image_descent import ImageDescent, FunctionDescent2D, surfaces
# you can put path to an image or a numpy array / torch tensor.
# It will be converted into black-and-white channel-last.
image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"

# coords are the initial coordinate to start optimization from
# either pixel location (int), or relative coordinates in (-1,-1) to (1,1) range (float).
descent = ImageDescent(image, coords=(915, 500))

optimizer = RandomBestGrad(descent.parameters(), lr = 1e-2, mul=1, bestof=10, opt = torch.optim.SGD(descent.parameters(), lr=1, momentum=0.9, nesterov=True))
torch.set_grad_enabled(False)
for i in range(100):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd)
descent.plot_path()
```

```python
import torch
from torchzero.optim import RandomGrad, RandomBestGrad
from image_descent import ImageDescent, FunctionDescent2D, surfaces
# you can put path to an image or a numpy array / torch tensor.
# It will be converted into black-and-white channel-last.
image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"

# coords are the initial coordinate to start optimization from
# either pixel location (int), or relative coordinates in (-1,-1) to (1,1) range (float).
descent = ImageDescent(image, coords=(915, 500))

optimizer = RandomBestGrad(descent.parameters(), lr = 1e-2, mul=1, bestof=10, 
                           opt = torch.optim.SGD(descent.parameters(), lr=2, momentum=0.9, nesterov=True), avgbest=False, negate_worst=True)
torch.set_grad_enabled(False)
for i in range(100):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd)
descent.plot_path()
```
