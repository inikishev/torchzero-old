```python
import torch
from torchzero.optim import RandomSearch, RandomWalk
from torchzero.benchmarks.restoration import Restoration
```

```python
benchmark = Restoration()
benchmark.evaluate(RandomSearch(benchmark.parameters()))
```

```python
from glio.plot import qlinechart
qlinechart(benchmark.loss_history)
```

```python
benchmark = Restoration()
benchmark.evaluate(RandomWalk(benchmark.parameters(),  1e-1))
```

```python
from glio.plot import qlinechart
qlinechart(benchmark.loss_history)
```
