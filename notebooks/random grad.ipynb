```python
import torch
from torchzero.optim import RandomGrad, RandomBestGrad
from image_descent import ImageDescent

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

# RandomGrad
Use random petrubation as gradients for any gradient-based optimizer

```python
descent = ImageDescent(image, coords=(915, 500))

optimizer = RandomGrad(descent.parameters(), lr=1e-4, opt=torch.optim.SGD(descent.parameters(), lr=5e-2))
for i in range(1000):
    print(i, end='\r')
    def closure():
        optimizer.zero_grad()
        return descent.step_nograd()
    optimizer.step(closure)
descent.plot_path()
```
