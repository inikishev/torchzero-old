```python
import torch
from torchzero.optim.zeroth_order.simulated_annealing import SimulatedAnnealing
from image_descent import ImageDescent, FunctionDescent2D, surfaces
from torchzero.random.random import Uniform

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

```python
descent = ImageDescent(image, coords=(915, 500))

optimizer = SimulatedAnnealing(descent.parameters(), lr=1e-1, init_temp=0.1, cooling_steps=500)
for i in range(1000):
    #print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```
