```python
import torch
from torchzero.optim import RandomOptimizer
from image_descent import ImageDescent, FunctionDescent2D, surfaces
from torchzero.random.random import Uniform

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

```python
descent = ImageDescent(image, coords=(915, 500))

optimizer = RandomOptimizer(descent.parameters(), lr=1e-1, stochastic=False, step_back=False)
for i in range(1000):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```

```python
descent = ImageDescent(image, coords=(915, 500))

optimizer = RandomOptimizer(descent.parameters(), lr=1e-1, stochastic=False, step_back=False, set_grad=True, opt=torch.optim.SGD(descent.parameters(), 1.))
for i in range(1000):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```
