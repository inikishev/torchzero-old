```python
import torch
from torchzero.optim.first_order import SoftSignGD
from torchzero.optim.first_order.gd import GD
from image_descent import ImageDescent, FunctionDescent2D, surfaces
from glio.plot import *

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

```python
descent = ImageDescent(image, coords=(915, 500))

optimizer = SoftSignGD(descent.parameters(), 5e-1)
for i in range(1000):
    print(i, end='\r')
    optimizer.zero_grad()
    descent.step()
    optimizer.step()
descent.plot_path()
```

```python
descent = FunctionDescent2D(surfaces.sphere)
optimizer = SoftSignGD(descent.parameters(), lr = 5)
for i in range(100):
    loss = descent.step()
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

descent.plot_path()
```
