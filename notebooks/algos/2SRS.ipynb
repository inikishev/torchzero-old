```python
import torch
from torchzero.optim.two_step_random_search import TwoStepRS
from image_descent import ImageDescent, FunctionDescent2D, surfaces

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

```python
descent = ImageDescent(image, coords=(915, 500))

optimizer = TwoStepRS(descent.parameters(), lr=1e-2, lr2=1e-1)
for i in range(1000):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```
