```python
import torch
from torchzero.optim import RandomSearch, RandomShrinkingSearch
from image_descent import ImageDescent

import matplotlib.pyplot as plt
image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

# Its just random search

```python
descent = ImageDescent(image, coords=(915, 500))
losses = []
optimizer = RandomSearch(descent.parameters())
for i in range(500):
    print(i, end='\r')
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```

```python
descent.plot_losses()
```

```python
def randlike_in_range(param,**kwargs):
    rand = torch.rand_like(param)
    return rand * 2 - 1
    
descent = ImageDescent(image, coords=(915, 500))

optimizer = RandomSearch(descent.parameters(), randlike_in_range)
for i in range(500):
    print(i, end='\r')
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```

# Random shrinking search
searches in a shrinking area around best parameters

```python
descent = ImageDescent(image, coords=(915, 500))
losses = []
optimizer = RandomShrinkingSearch(descent.parameters(), lr=0.5, nsteps=500)
for i in range(500):
    print(i, end='\r')
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```

```python
descent.plot_losses()
```
