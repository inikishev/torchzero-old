```python
import torch
from torchzero.optim.zeroth_order.grid_search import SequentialSearch
from image_descent import ImageDescent, FunctionDescent2D, surfaces

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

```python
descent = ImageDescent(image, coords=(900, 400))

optimizer = SequentialSearch(descent.parameters(), step = 2e-2, domain = [-1, 1])
for i in range(1000):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd_search)
    descent.step_nograd()
descent.plot_path()
```
