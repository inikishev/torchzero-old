```python
import torch
from torchzero.optim import FractionalOptimizer
from image_descent import ImageDescent, FunctionDescent2D, surfaces

import matplotlib.pyplot as plt
image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

# Vanilla SGD

```python
descent = FunctionDescent2D(surfaces.rosenbrock)
optimizer = torch.optim.SGD(descent.parameters(), 1)
for i in range(500):
    print(i, end='\r')
    optimizer.zero_grad()
    loss = descent.step()
    loss.backward()
    optimizer.step()
descent.plot_path()
```

# 1.75th order SGD

```python
descent = FunctionDescent2D(surfaces.rosenbrock)
optimizer = FractionalOptimizer(descent.parameters(), v=1.75, opt = torch.optim.SGD(descent.parameters(), 2e-1))
for i in range(500):
    print(i, end='\r')
    optimizer.zero_grad()
    loss = descent.step()
    loss.backward()
    optimizer.step()
descent.plot_path()
```
