```python
descent = FunctionDescent2D(surfaces.convex)
optimizer = QuadraticLS(descent.parameters(), lr = 0.1)
for i in range(2):
    def closure(backward:bool):
        loss = descent.step()
        if backward: 
            optimizer.zero_grad()
            loss.backward()
        return loss
    optimizer.step(closure)
    loss = descent.step()
    print(f'{loss = }')
descent.plot_path()
```
