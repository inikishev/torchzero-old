```python
import optuna
import torch
from torchzero.optim.integrations.optuna import OptunaOptimizer
from image_descent import ImageDescent, FunctionDescent2D, surfaces

image = r"F:\Stuff\Programming\AI\pytorch image descent\surfaces\spiral.jpg"
```

```python
descent = ImageDescent(image, coords=(915, 500))

sampler = optuna.samplers.RandomSampler()
optimizer = OptunaOptimizer(descent.parameters(), sampler = sampler, domain = (-1, 1))
for i in range(1000):
    print(i, end='\r')
    optimizer.zero_grad()
    optimizer.step(descent.step_nograd_search)
descent.plot_path()
```
