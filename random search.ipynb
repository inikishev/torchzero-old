```python
import torch
from torchzero.benchmarks.restoration import Restoration
from torchzero.optim.random_search import RandomSearch
```

```python
model = Restoration(torch.tensor([1., 2., 3,]))
opt = RandomSearch(model.parameters())
for _ in range(1000):
    
    opt.step(model)

import matplotlib.pyplot as plt
plt.plot(model.loss_history)
```
