```python
import torch
from torchzero.benchmarks.restoration import Restoration
from torchzero.optim.random_walk import RandomWalk
```

```python
model = Restoration(torch.tensor([1., 2., 3,]),loss=torch.nn.L1Loss(), init=torch.zeros_like)
opt = RandomWalk(model.parameters(), lr=1e-2, momentum=0.1)
for i in range(10000):
    print(i, end='\r')
    opt.step(model)

import matplotlib.pyplot as plt
plt.plot(model.loss_history)
```
